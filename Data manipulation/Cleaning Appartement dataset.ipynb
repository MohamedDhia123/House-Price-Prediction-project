```python
import pandas as pd
df = pd.read_csv("New_Appartements.csv")

df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)

df.drop(columns="Pieces", inplace=True)
df.drop(columns="Age", inplace=True)
df.drop(columns="uknown", inplace=True)
df.drop(columns="Stage", inplace=True)
df.head(20)
```

```python
df["Location"] = df["Location"].str.replace("inLa Marsa","")
df["Location"] = df["Location"].str.replace("inNabeul","")
df["Location"] = df["Location"].str.replace("inLe Kram","")
df["Location"] = df["Location"].str.replace("inHammamet","")
df["Location"] = df["Location"].str.replace("inCarthage","")
df["Location"] = df["Location"].str.replace("inAriana Ville","")
df["Location"] = df["Location"].str.replace("inSousse Ville","")
df["Location"] = df["Location"].str.replace("inLa Soukra","")
df["Location"] = df["Location"].str.replace("inSousse Jaouhara","")
df["Location"] = df["Location"].str.replace("inLe Bardo","")
df["Location"] = df["Location"].str.replace("inHammam Sousse","")
df["Location"] = df["Location"].str.replace("inLa Manouba","")
df["Location"] = df["Location"].str.replace("inEl Menzah","")
df["Location"] = df["Location"].str.replace("inCité El Khadra","")
df["Location"] = df["Location"].str.replace("inAkouda","")
df["Location"] = df["Location"].str.replace("inRaoued","")
df["Location"] = df["Location"].str.strip()
```

```python
df["Price"] = df["Price"].str.replace("TND","")
df["Price"] = df["Price"].str.replace("EUR","")
df["Price"] = df["Price"].str.replace("Decreased by200","")
df["Price"] = df["Price"].str.replace("Decreased by500 ","")
df["Price"] = df["Price"].str.replace("Decreased by50","")
df["Price"] = df["Price"].str.replace("Decreased by90","")
df["Price"] = df["Price"].str.replace("Decreased by300","")
df["Price"] = df["Price"].str.replace("Decreased by100","")
df["Price"] = df["Price"].str.replace("Decreased by250","")
df["Price"] = df["Price"].str.replace("Decreased by150","")
df["Price"] = df["Price"].str.replace("Decreased by170","")
df["Price"] = df["Price"].str.replace("Decreased by30","")
df["Price"] = df["Price"].str.replace("Decreased by600","")
df["Price"] = df["Price"].str.replace("Decreased by400","")
df["Price"] = df["Price"].str.replace(",","")
df["Price"] = df["Price"].str.strip()
```

```python
df["Rooms"] = df["Rooms"].str.replace("Rooms","")
df["Rooms"] = df["Rooms"].str.replace("Room","")
df["Rooms"] = df["Rooms"].str.strip()
```

```python
df["Area"] = df["Area"].str.replace("m²","")
df["Area"] = df["Area"].str.strip()
```

```python
df["Bathrooms"] = df["Bathrooms"].str.replace("Bathrooms","")
df["Bathrooms"] = df["Bathrooms"].str.replace("Bathroom","")
df["Bathrooms"] = df["Bathrooms"].str.strip()
```

```python
df['Condition'].fillna(method='ffill', inplace=True) 
FeaturesToBeComplete = ['Sea views', 'Washing machine', 'TV', 'Mountains views', 'Reinforced Door',
                        'Air conditioning', 'Garden', 'Heating', 'Fridge', 'Microwave', 'Concierge',
                        'Pool', 'Garage', 'Flooring', 'Internet', 'Oven', 'Elevator', 'Satellite dish',
                        'Terrace', 'Security system', 'Double glazing', 'Equipped kitchen', 'Furnished']

for feature in FeaturesToBeComplete:
    if df[feature].isnull().all:
        df[feature] = df[feature].fillna("no")

df = df.drop(6)

```

```python
df["Price"] = df["Price"].str.replace(" ","")
df = df.dropna(subset=['Price'])
df.head(30)
```
