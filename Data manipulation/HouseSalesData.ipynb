```python
import pandas as pd
import numpy as np

df = pd.read_csv("HouseDataSet.csv")

df.drop_duplicates(inplace=True)

df.drop(columns="Floor", inplace=True)
df.drop(columns="Pieces", inplace=True)

df.head(20)
```

```python
df.info()
```

```python

if df['Location'].str.contains(' in').any():
    df[['City', 'Region']] = df['Location'].str.split(' in', expand=True)
else:
    df['City'] = df['Location']
    df['Region'] = df['Location']

df["Location"] = df["Location"].str.strip()
```

```python
df.loc[df['Region'].isnull(), 'Region'] = df['Location']
```

```python
df['Price'] = df['Price'].str.replace(r'\D', '', regex=True)
df["Price"] = df["Price"].str.replace(",","")
df["Price"] = df["Price"].str.replace("1300000100000","1300000")
df["Price"] = df["Price"].str.strip()
df = df.dropna(subset=['Price'])
```

```python
df['Condition'].fillna(method='ffill', inplace=True) 
```

```python
df.drop(columns="Age", inplace=True)
df.drop(columns="Location", inplace=True)
```

```python

df["Rooms"] = df["Rooms"].str.replace("Rooms","")
df["Rooms"] = df["Rooms"].str.replace("Room","")
df["Rooms"] = df["Rooms"].str.strip()
```

```python
df["Area"] = df["Area"].str.replace("m²","")
df["Area"] = df["Area"].str.strip()
```

```python
df["Bathrooms"] = df["Bathrooms"].str.replace("Bathrooms","")
df["Bathrooms"] = df["Bathrooms"].str.replace("Bathroom","")
df["Bathrooms"] = df["Bathrooms"].str.strip()
```
