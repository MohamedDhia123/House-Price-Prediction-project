```python
import pandas as pd
import numpy as np


df = pd.read_csv("cleaned_dataset.csv")
df.head()
```

```python
df = df.dropna(subset=['Price'])
df = df.dropna(subset=['Area'])
df = df.dropna(subset=['Rooms'])
df = df.dropna(subset=['Bathrooms'])
```

```python
df["Price"] = df["Price"].str.replace("Decreasedby1000","")
df["Price"] = df["Price"].str.replace("Decreasedby3000","")
df["Price"] = df["Price"].str.replace("Decreasedby1000","")
df["Price"] = df["Price"].str.replace("Decreasedby1500","")
df["Price"] = df["Price"].str.replace("Decreasedby800","")
df["Price"] = df["Price"].str.replace("Decreasedby700","")
df["Price"] = df["Price"].str.replace("Decreasedby40","")
```

```python
df['Price'] = df['Price'].astype(float)
df['Area'] = df['Area'].astype(int)
df['Rooms'] = df['Rooms'].astype(int)
df['Bathrooms'] = df['Bathrooms'].astype(int)
df[df['Price'] > 20000]
```

```python
df = df[df['Price'] <= 10000]
df = df[df['Area'] <= 5000]


df = df[(df['Type'] != 'apartment') | ((df['Type'] == 'apartment') & (df['Area'] <= 5000))]


df.shape
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
